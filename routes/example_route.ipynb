```python
import os
import pandas as pd

DATABASE = 'ant1_route3'
df = pd.read_csv(os.path.join(DATABASE, 'database_entries.csv'))

# strip whitespace from column names
df = df.rename(columns=lambda x: x.strip())
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(df['X [mm]'])
y = np.array(df['Y [mm]'])
bearings = np.deg2rad(df['Heading [degrees]'])

_, ax = plt.subplots()
ax.axis('equal')
ax.plot(x, y)

# show start and end of route
ax.plot(x[0], y[0], 'g.', x[-1], y[-1], 'r.', markersize=15)

# plot arrows along route
STEP = 100
x = x[::STEP]
y = y[::STEP]
bearings = bearings[::STEP]
u = np.sin(bearings)
v = np.cos(bearings)
ax.quiver(x, y, u, v, zorder=3);
```
